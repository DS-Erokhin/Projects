# Исследование объявлений о продаже квартир

Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

## Предобработка данных

### Замена пропущенных значений

Перед тем, как переходить к замене пропущенных значений необходимо определиться в каких столбцах каким способом заполнять пропуски. Для этого по подробнее ознакомимся с данными.

```python
df
```

```python
df = df.drop(df[df['floors_total']==0].index)
df = df.drop(df[df['locality_name']==0].index)

df[['ceiling_height', 'floors_total','living_area', 'balcony']] = df[['ceiling_height', 'floors_total','living_area', 'balcony']].fillna(0)

#Замена пропущенных значений в столбце 'ceiling_height'
ceiling_height_median = df[(df['ceiling_height']>2)&(df['ceiling_height']<4)]['ceiling_height'].median()
print('Медиана высоты потолков:', ceiling_height_median)
print('')

df.loc[(df['ceiling_height']<2)|(df['ceiling_height']>4), 'ceiling_height']=ceiling_height_median

#Замена пропущенных значений в столбце 'living_area'
living_area_median = df[df['living_area']>10]['living_area'].median()
print('Медиана метража жилой площади:', living_area_median)

df.loc[df['living_area']<10,'living_area']=living_area_median
```

#### Вывод
Были заменены пропущенные значения в столбцах 'ceiling_height', 'floors_total','living_area', 'kitchen_area', 'balcony'. Не тронутыми остались столбцы 'days_exposition', 'ponds_nearest', 'ponds_around3000', 'parks_nearest', 'parks_around3000', 'cityCenters_nearest', 'airports_nearest'. Решение не заполнять пропуски принято в связи с тем, что пропуски могут означать отсутствие объекта рядом с недвижимостью.

### Замена типа данных

```python
#Замена типа данных в столбце "is_apartment"
df['is_apartment']=df['is_apartment'].astype('bool')


#Замена типа данных в столбце "first_day_exposition"
df['first_day_exposition']=pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')


#Замена типа данных в столбцах "floors_total", "last_price", "total_area", "balcony" и "living_area"
df[['floors_total', 'last_price', 'total_area', 'balcony', 'living_area']]=df[['floors_total', 'last_price', 'total_area', 'balcony', 'living_area']].astype('int')


df['kitchen_area'] = round(df['kitchen_area']).astype('Int64')
df[['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 
    'ponds_around3000', 'ponds_nearest', 'days_exposition']] = round(df[['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 
    'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'days_exposition']]).astype(pd.Int32Dtype())

df.info()
```

#### Вывод
Были заменены типы данных в столбцах на int, в столбце 'first_day_exposition' на 'datetime' и в столбце 'is_apartment' на bool, поскольку это фиктивная переменная.

## Расчёты и добавление результатов в таблицу

```python
def floor(row):
    if row['floor']==1:
        return 'первый'
    if row['floor']==row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
df['floor_category'] = df.apply(floor, axis=1)
```

```python
#Цена за квадратный метр
df['price_per_square_meter'] = df['last_price']/df['total_area']

#Доля жилой площади
df['living_area_share'] = df['living_area']/df['total_area']

#Доля кухни
df['kitchen_area_share'] = df['kitchen_area']/df['total_area']
```

#### Вывод
Добавлены в основной датафрейм столбцы 'price_per_square_meter', 'floor_category', 'living_area_share' и 'kitchen_area_share'.

## Исследовательский анализ данных

### Изучение времени продажи квартиры

```python
days_exposition_mean = df['days_exposition'].mean()
print('Среднее время продажи квртиры:', round(days_exposition_mean), 'день')

days_exposition_median = df['days_exposition'].median()
print('Медианное время продажи квртиры:', round(days_exposition_median), 'день')
```

Для изучения времени продажи квартиры возьмем те объекты, в которых объявление весело не более 1000 дней. Данное значение было взято условно и может варьироваться. Все, что продавалось более 1000 дней будем считать выбросами и не учитывать в анализе.

```python
df['days_exposition_good'] = df[df['days_exposition']<1000]['days_exposition']
```

Посчитаем среднее и медиану для нового столбца.

```python
days_exposition_good_mean = df['days_exposition_good'].mean()
print('Среднее время продажи квртиры:', round(days_exposition_good_mean), 'день')

days_exposition_good_median = df['days_exposition_good'].median()#93
print('Медианное время продажи квртиры:', round(days_exposition_good_median), 'день')

df['days_exposition'].hist(bins=50, figsize=(7, 5))
plt.show()
print(df['days_exposition'].value_counts().head(20))
```

```python
print(df[df['days_exposition']<7]['days_exposition'].value_counts()) #продаваться недвижимость начинает с 3 дня, в 1 и 2 день
#продажи практически отсутствуют, 1 продажа в первый день и 3 во второй, в третий же уже 158
```

#### Вывод
Замечены странные значения. Продаваться недвижимость начинает с 3 дня, в 1 и 2 день продажи практически отсутствуют, возможно, это связано с алгоритмами выдачи объявлений на площадке, либо 1-2 просто минимальный срок для продажи квартиры по выгодной цене. Также, замечено, что много квартир продается в первые ~20 дней и, присутствуют пики продаж недвижимости на 30, 45, 60 и 90 день.

```python
#Зависимость цены от удаленности от центра
print(df[['last_price_good', 'cityCenters_nearest']].corr()) #Корреляция = -0.35. Отрицательная корреляция логична, поскольку
#чем дальше от центра, тем дешевле недвижимость
df.plot(kind='scatter', x='cityCenters_nearest', y='last_price_good', grid=True, sharex=False, alpha=0.03, title='Зависимость цены от удаленности от центра');
```

```python
#Зависимость цены от этажа
round(df.pivot_table(index='floor_category', values='last_price_good')) #
```

Видна зависимость цены от этажа, если рассматривать по категориям. Так, самый дорогие квартиры располагаются на средних этажах, а самые дешевые на первом.

```python
#Зависимость цены от даты размещения
print(df[['last_price_good', 'days_exposition_good']].corr()) 
print('---------------------------------------------------')

df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
print(df[['last_price_good', 'weekday', 'month']].corr())
```

Корреляция отсутствует. Цена продажи практически не зависит от того, сколько дней висело объявление.

```python
df.pivot_table(index='weekday', values = 'last_price_good', aggfunc = ['mean', 'count']) 
```

По субботам и воскресеньям заключается меньше всего сделок купли-продажи, больше всего в середине недели, со вторника по пятницу

```python
pivot = round(df.pivot_table(index='month', values = 'last_price_good', aggfunc = ['mean', 'count']))
plt.figure(figsize = (10, 5))
plt.title('Кол-во проданных квартир в каждом месяце')
pivot['count']['last_price_good'].plot(style='o-');
pivot
```

Наиболее активными периодами, когда люди покупают недвижимость является промежуток с февраля по апрель (включительно), с мая по сентябрь квартир покупается меньше всего, однако под конец года кол-во покупаемых квартир снова увеличивается и достигает пика в ноябре.

```python
pivot = round(df.pivot_table(index='year', values = 'last_price_good', aggfunc = ['mean', 'count']))

plt.figure(figsize = (10, 5))
plt.title('Общее количество проданных квартир от года к году')
pivot['count']['last_price_good'].plot(style='o-');

plt.figure(figsize = (10, 5))
plt.title('Как менялась средняя цена проданных квартир от года к году')
pivot['mean']['last_price_good'].plot(style='o-');

pivot
```

Сразу бросается в глаза, что в 2014 году продано всего 124 квартиры, можно предположить, что столь малое кол-во, например, может быть связано что сервис по мониторингу проданной недвижи начал свою работу только в 2014 году, точнее в его конце. С чем же связан бум в 2017-2018 годах предположить сложнее. Также, можно заметить как с увеличением количества проданных квартир уменьшается их цена.

#### Вывод
Больше всего стоимость жилья зависит от:
1. Площади. Чем больше площадь, тем выше и цена.
2. Количества комнат. Чем больше комнат, тем выше цена.
3. Этажа. На первом этаже квартиры самые дешевые.
4. Дня недели. В будни недвижимость покупается чаще.

Доля больших квартир в центре меньше, чем по городу.
